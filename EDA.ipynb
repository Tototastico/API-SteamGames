Vamos a hacer un peque√±o EDA sobre el dataset de juegos, ya que necesitamos conocer un poco mas nuestros datos.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df_games = pd.read_parquet('clean_games.parquet.gzip') # Leemos el archivo
df_games.head() #Damos un vistazo rapido de los juegos
```

```python
df_games.groupby('genres').count()['id']
```

```python
# Configura el tama√±o del gr√°fico si es necesario
plt.figure(figsize=(10, 6))

# Crea el gr√°fico de barras
conteo_genero = df_games['genres'].value_counts().sort_values(ascending=False)
conteo_genero.plot(kind='bar',color='skyblue')

# Personaliza los ejes y el t√≠tulo
plt.xlabel('G√©nero')
plt.ylabel('Cantidad de Registros')
plt.title('Cantidad de Registros por G√©nero')

# Muestra el gr√°fico
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.boxplot(x='price', data=df_games)
plt.ylabel('Precio')
plt.title('Outlier de precios')
```

```python
empresas = df_games['publisher'].value_counts().sort_values(ascending=False).to_dict()
empresas = dict(list(empresas.items())[:10])
```

```python
etiquetas = list(empresas.keys())
valores = list(empresas.values())

# Configura el tama√±o del gr√°fico si es necesario
plt.figure(figsize=(10, 6))
sns.barplot(x=etiquetas, y=valores, palette="viridis")
# Personaliza los ejes y el t√≠tulo
plt.xlabel('G√©nero')
plt.ylabel('Cantidad de Registros')
plt.title('Cantidad de Registros por G√©nero')

plt.xticks(rotation=45)

# Muestra el gr√°fico
plt.show()
```

```python
# Configura el tama√±o del gr√°fico si es necesario
plt.figure(figsize=(10, 6))
sns.lineplot(df_games['year'].value_counts(), palette="viridis")
# Personaliza los ejes y el t√≠tulo
plt.xlabel('A√±o')
plt.ylabel('Cantidad de juegos')
plt.title('Cantidad de juegos por a√±o')

plt.xticks(rotation=45)

# Muestra el gr√°fico
plt.show()
```
